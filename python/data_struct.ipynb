# LinkedLists

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:

    def __init__(self):
        self.head = None    

    # O(n)
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return 
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    # O(n)
    def find(self, key):
        current = self.head
        while current is not None:
            if current.data == key:
                return True
            current = current.next            
        return False

    # O(n)
    def delete(self, key):
        current = self.head
        if current and current.data == key:
            self.head = current.next
            current = None
            return

        prev = None
        while current is not None:
            if current.data == key:
                break
            prev = current
            current = current.next

        if current is None:
            return

        prev.next = current.next
        current = None    

    def to_list(self):
        result = []
        current = self.head
        while current:
            result.append(current.data)
            current = current.next
        return result        
```

```python
ll = LinkedList()

## APPEND TESTS
ll.append(1)
assert ll.to_list() == [1], f"should be [1]"

ll.append(2)
ll.append(3)
assert ll.to_list() == [1, 2, 3], f"should be [1, 2 ,3]"

# FIND TESTS
assert ll.find(2), f"Should find 2"
assert ll.find(3), f"Should find 3"
assert ll.find(4) == False, f"Should find 4"

## DELETE
ll.delete(2)
assert ll.to_list() == [1,3], f"should be [1,3] after delete"
ll.delete(1)
assert ll.to_list() == [3], f"should be [3] after delete"
ll.delete(3)
assert ll.to_list() == [], f"should be [] after delete"
```

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

Example 1:

Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]

Example 2:
Input: list1 = [], list2 = []
Output: []

Example 3:
Input: list1 = [], list2 = [0]
Output: [0]
 

Constraints:

The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:

        # create a sentinel or initial Node        
        dummy = ListNode()
        # set the reference
        tail = dummy

        # compare the both inputs 1 < 2 
        while list1 and list2:
            if list1.val <  list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next

        if list1:
            tail.next = list1
        else:
            tail.next = list2

        return dummy.next
```
